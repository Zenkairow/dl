```python
from ultralytics import YOLO
import cv2
import numpy as np
```

```python
model = YOLO("yolov8n.pt")
```

```python
import urllib.request

url = "https://ultralytics.com/images/bus.jpg"
urllib.request.urlretrieve(url, "sample.jpg")
```

```python
img = cv2.imread("sample.jpg")

results = model(img)

annotated = results[0].plot()

cv2.imshow("Image Detection", annotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
vehicle_classes = [2, 3, 5, 7]
person_class = 0

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)[0]

    boxes = results.boxes

    filtered_boxes = []
    for box in boxes:
        cls = int(box.cls[0])
        if cls in vehicle_classes or cls == person_class:
            filtered_boxes.append(box)

    results.boxes = filtered_boxes

    annotated = results.plot()

    cv2.imshow("Filtered Detection", annotated)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
import urllib.request

model = YOLO("yolov8n.pt")

url = "https://ultralytics.com/images/zidane.jpg"
urllib.request.urlretrieve(url, "vehicle2.jpg")

img = cv2.imread("vehicle2.jpg")

results = model(img)

annotated = results[0].plot()

cv2.imshow("Vehicle Detection", annotated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
