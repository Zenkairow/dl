```python
%pip install -q kagglehub
import kagglehub, os, glob

path = kagglehub.dataset_download("lakshmi25npathi/imdb-dataset-of-50k-movie-reviews")
csv_files = glob.glob(os.path.join(path, '**', '*.csv'), recursive=True)
CSV_PATH = csv_files[0]
print("Using:", CSV_PATH)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(CSV_PATH)
df = df.sample(n=4000, random_state=42).reset_index(drop=True)
```

```python
df
```

```python
df['sentiment'].value_counts()
```

```python
map={'positive':1,'negative':0}
df['sentiment']=df['sentiment'].map(map).astype(int)
```

```python
df
```

```python
train_data_len=len(df)*0.80
```

```python
test_data_len=len(df)-train_data_len
```

```python
train_data = df.iloc[0:int(train_data_len), :]  
test_data = df.iloc[int(train_data_len):, :] 
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
vectorize_layer = layers.TextVectorization(
    max_tokens=30,
    output_mode="int",
)
```

```python
vectorize_layer.adapt(df['review'].values)
X = vectorize_layer(train_data['review'].values)
y = train_data['sentiment'].values
```

```python
model_lstm = tf.keras.Sequential([
    tf.keras.layers.Embedding(10000, 128),
    tf.keras.layers.LSTM(64),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_lstm.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model_lstm.fit(X, y, epochs=5, batch_size=32)
```

```python
model_gru = tf.keras.Sequential([
    tf.keras.layers.Embedding(10000, 128),
    tf.keras.layers.GRU(64),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model_gru.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model_gru.fit(X, y, epochs=5, batch_size=32)
```
